## 2. Configure the benchmark

Change `MODE` below to control how long the benchmark runs.  
Start with `"smoke"` for a quick demo (~15 min total).

```python
# ──────────────────────────────────────────────
# ⚙️  Configuration — edit these if you like
# ──────────────────────────────────────────────
MODE  = "smoke"                          # "smoke" | "medium" | "full"
ALGOS = "random dqn dqn_paper ppo"       # space-separated subset, or leave all
SEED  = 0
```

## 3. Run the benchmark 🚀

```python
import subprocess, sys, os

os.chdir("/content/repo/task 9")

cmd = [
    sys.executable, "rl_benchmark.py",
    "--mode", MODE,
    "--seed", str(SEED),
    "--algos", *ALGOS.split(),
]

print(f"Running: {' '.join(cmd)}")
print("=" * 60)
result = subprocess.run(cmd, text=True)
print("\n✅ Benchmark finished!" if result.returncode == 0 else f"\n❌ Exited with code {result.returncode}")
```

## 4. View the results

```python
import glob
from IPython.display import display, Image, Markdown
import pandas as pd

# Find the most recent results folder
results_dirs = sorted(glob.glob("results/benchmark_*"))
if not results_dirs:
    print("No results found — did the benchmark run?")
else:
    out = results_dirs[-1]
    print(f"📁 Results directory: {out}\n")

    # Leaderboard
    csv_path = os.path.join(out, "leaderboard.csv")
    if os.path.exists(csv_path):
        display(Markdown("### 🏆 Leaderboard"))
        display(pd.read_csv(csv_path))

    # Plots
    for name in ["reward_curves.png", "loss_curves.png", "policy_distribution.png", "ppo_components.png"]:
        img_path = os.path.join(out, name)
        if os.path.exists(img_path):
            display(Markdown(f"### {name.replace('_', ' ').replace('.png', '').title()}"))
            display(Image(filename=img_path, width=700))
```
